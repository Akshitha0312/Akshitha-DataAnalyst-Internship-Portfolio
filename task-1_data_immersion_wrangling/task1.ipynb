```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\\Users\\prash\\OneDrive\\docs\\Documents\\Desktop\\ApexPlanet_DataAnalytics_Dataset.xlsx")
```

```python
# Data familiarization

print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nFirst 5 Rows:")
display(df.head())

print("\nLast 5 Rows:")
display(df.tail())

print("\nDataset Info:")
df.info()
```

```python
# Data Quality Assessment

print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
#Summary Statistics
df.describe(include='all')
```

```python
# Remove Duplicates

df = df.drop_duplicates()

print("Duplicates Removed")
print(df.shape)
```

```python
# Covert date format

df['Order_Date'] = pd.to_datetime(df['Order_Date'])

print(df.dtypes)
```

```python
# Handle missing values

#df.fillna("Unknown", inplace=True)

print(df.isnull().sum())
```

```python
# Outlier detection

Q1 = df['Total_Sales'].quantile(0.25)

Q3 = df['Total_Sales'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[(df['Total_Sales'] < (Q1 - 1.5*IQR)) |
              (df['Total_Sales'] > (Q3 + 1.5*IQR))]

print("Number of Outliers:", len(outliers))
```

```python
df['Year'] = df['Order_Date'].dt.year

df['Month'] = df['Order_Date'].dt.month_name()

df['Day'] = df['Order_Date'].dt.day_name()

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0,25,40,60,100],
    labels=['Young','Adult','Middle Age','Senior']
)

display(df.head())
```

```python
# Data dictionary

data_dictionary = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes.astype(str)
})

display(data_dictionary)
```

```python
# Export files

df.to_csv("cleaned_dataset.csv", index=False)

data_dictionary.to_csv("data_dictionary.csv", index=False)

print("Files Exported Successfully")
```
